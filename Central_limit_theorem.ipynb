# Central Limit Theorem

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

## Uniform Distribution

```python
# Set the parameters
num_samples=1000
sample_size=300
distribution_range=(0,1)
```

```python
# Generate samples from a uniform distribution
samples=np.random.uniform(distribution_range[0],distribution_range[1],(num_samples,sample_size))
```

```python
samples.shape
```

```python
# Calculate the samples mean
samples_mean=np.mean(samples,axis=1)
```

```python
# Plot the histogram of the sample mean
plt.hist(samples_mean,bins=25,density=True,edgecolor='black')
plt.title('Histogram of sample means')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.show()
```

```python
# Plot the histogram of the sample means
sns.kdeplot(samples_mean)
```

```python
# Set the parameters
num_samples=2000
sample_size=200
lambda_param=2
```

```python
# Generate samples from a exponential distribution
samples=np.random.exponential(scale=1/lambda_param,size=(num_samples,sample_size))
```

```python
# Calculate the samples mean
sample_means=np.mean(samples,axis=1)
```

```python
# Plot the histogram of the sample mean
plt.hist(sample_means,bins=25,density=True,edgecolor='black')
plt.title('Histogram of sample means')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.show()
```

```python
# Plot the kernel density estimation of the sample means
sns.kdeplot(sample_means)
```

```python
# Set the parameters
num_samples=1000
sample_size=50

# poisson Distribution parameter
poisson_lambda=5

# Binomial Distribution parameter
binomial_n=10
binomial_p=0.5

# Gamma Distribution parameter
gamma_shape=2
gamma_scale=2
```

```python
# Generate samples from a poisson distribution
poisson_samples=np.random.poisson(lam=poisson_lambda,size=(num_samples,sample_size))

# Generate samples from a binomial distribution
binomial_samples=np.random.binomial(n=binomial_n,p=binomial_p,size=(num_samples,sample_size))

# Generate samples from a gamma distribution
gamma_samples=np.random.gamma(shape=gamma_shape,scale=gamma_scale,size=(num_samples,sample_size))
```

```python
# Calculate the samples mean
poisson_sample_mean=np.mean(poisson_samples,axis=1)
binomial_sample_mean=np.mean(binomial_samples,axis=1)
gamma_sample_mean=np.mean(gamma_samples,axis=1)
```

```python
# Plot the histograms of sample means
fig,axs=plt.subplots(3,1,figsize=(10,15))

# Poisson Distribution
axs[0].hist(poisson_sample_mean,bins=30,density=True,edgecolor='black')
axs[0].set_title('Histograms of Sample means(Poisson Distribution)')
axs[0].set_xlabel('sample Mean')
axs[0].set_ylabel('Density')

# Binomial Distribution
axs[1].hist(binomial_sample_mean,bins=30,density=True,edgecolor='black')
axs[1].set_title('Histograms of Sample means(Binomial Distribution)')
axs[1].set_xlabel('sample Mean')
axs[1].set_ylabel('Density')

# Gamma Distribution
axs[2].hist(gamma_sample_mean,bins=30,density=True,edgecolor='black')
axs[2].set_title('Histograms of Sample means(Gamma Distribution)')
axs[2].set_xlabel('sample Mean')
axs[2].set_ylabel('Density')
```

```python
# Plot the kernel density estimation of the sample means of Poisson Distribution
sns.kdeplot(poisson_sample_mean)
```

```python
# Plot the kernel density estimation of the sample means of Binomial Distribution
sns.kdeplot(binomial_sample_mean)
```

```python
# Plot the kernel density estimation of the sample means of Gamma Ditribution
sns.kdeplot(gamma_sample_mean)
```
